```python
dataset = [2,4,56,78,2,3,4,56,8,9,0,12,45,]
sorted(dataset)
```

```python
#function for computing outliers from zscore
outliers = []
def detect_outliers(data):
    threshold = 3
    mean = np.mean(data)
    std = np.std(data)
    
for i in dataset:
    z_score = (i- mean)/std
    if np.abs(z_score)> threshold:
        outliers.append(i)
    
return outliers    
    
```

```python
detect_outliers(dataset)
print(detect_outliers)
```

1)sort the data

2)calculate Q1 and Q3

3)IQR (Q3- Q1)

4)Find The lower Fence (Q1-1.5) *IQR

5)Find The upper Fence (Q1+1.5) *IQR

```python
dataset = sorted(dataset)
q1,q3 =  np.percentile(dataset, [25,75])
print(q1,q3)
```

```python
iqr = q3-q1
print(iqr)
```

```python
#Find the lower fence nad higher fence
lower_fence = q1-(1.5*iqr)
higher_fence = q1+(1.5*iqr)
```

```python
print(lower_fence, higher_fence)
```
